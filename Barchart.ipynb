# Bar

creating bar

```python
import numpy as np
import matplotlib.pyplot as plt


x_points = np.array(['php','c++','java','html','css','javascript'])
y_points = np.array([2,4,6,8,10,12])

plt.bar(x_points, y_points, width=0.1,color = 'c')
plt.show()
```

Horizontal bars

```python
x_points = np.array(['php','c++','java','html','css','javascript'])
y_points = np.array([22,24,26,28,30,32])

plt.barh(x_points, y_points, height=0.1,color = 'y')
plt.show()
```
