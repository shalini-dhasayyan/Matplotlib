```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

                                    # generate some random data
data_1 = np.random.normal(0, 7, 78)
data_2 = np.random.normal(0, 9, 99)

data = list([data_1, data_2])
fig, ax = plt.subplots()
# build a box plot
ax.boxplot(data)


                                    # title and label axis
ax.set_title('weather')
ax.set_xlabel('city')
ax.set_ylabel('temperature')
xticklabels=['netherland', 'portuguese']
ax.set_xticklabels(xticklabels)


                      # adding horizontal grid lines
ax.yaxis.grid(True)

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

                                  # generate some random data
data_1 = np.random.normal(0, 4, 67)
data_2 = np.random.normal(0, 5, 78)

data = list([data_1, data_2])

fig, axis = plt.subplots()


                                         # build a violin plot
axis.violinplot(data, showmeans=False, showmedians=True)


# add title and axis labels
axis.set_title('weather')
axis.set_xlabel('city')
axis.set_ylabel('temperature')


                                                                       
                          # add x-tick labels
xticklabels = ['chiacago', 'arizona']
axis.set_xticks([1,2])
axis.set_xticklabels(xticklabels)


# add horizontal grid lines
axis.yaxis.grid(True)


# show the plot
plt.show()
```
