# Add grid lines

```python
plt.plot(x,color='b')
plt.plot(y,color = 'r')
plt.title("line")
plt.xlabel("count x")
plt.ylabel("count y")
plt.grid()
plt.show()
```

# Add linestyle & linewidth to grid
