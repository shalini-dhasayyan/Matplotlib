# Scatter plot              

Creating single scatter plot

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x_point = np.array([0,2,3,4,5,6,7,8,9,10])
y_point = np.array([12,14,16,18,19,20,22,24,26,28])
plt.scatter(x_point, y_point, color = 'y')
```

 multiple scatterplot

```python
x_point = np.array([0,2,3,4,5,6,7,8,9,10])
y_point = np.array([12,14,16,18,19,20,22,24,26,28])
plt.scatter(x_point, y_point, color = 'y')

x_point = np.array([3,6,7,4,7,3,8,8,9])
y_point = np.array([11,15,20,25,30,35,40,44,50])
plt.scatter(x_point, y_point, color = 'r')

x_point = np.array([3,6,7,4,7,3,8,8,9])
y_point = np.array([11,20,25,35,44,50,55,60,65])
plt.scatter(x_point, y_point, color ='b')
```

Add Cmap,sizes,alpha

```python
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, s=sizes, alpha=0.5,color='r')

plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='PuRd')

plt.colorbar()

plt.show()
```
