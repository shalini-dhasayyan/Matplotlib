# Subplot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
xpoint = np.array([0, 1, 2, 3])
ypoint = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(xpoint,ypoint,color='r')#add title,label,color
plt.title('line plot')

xpoint = np.array([2, 4, 5, 7])
ypoint = np.array([3, 8, 10, 12])
plt.title('line plot')
plt.xlabel('countx')
plt.ylabel('county')


plt.subplot(1, 2, 1)
plt.plot(xpoint,ypoint,color='g')#add title,label,color
plt.title('line plot')
plt.xlabel('countx')
plt.ylabel('county')
plt.show()
```
